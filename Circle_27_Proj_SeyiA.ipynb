```python
crime_count_by_type = crime_df.groupby("Primary Type")["ID"].sum()
```

```python
crime_count_by_type.head(2)
```

```python
crime_count_by_type.to_frame().head()
```

```python
plt.figure(figsize=(12, 6))
crime_count_by_type.sort_values().plot(kind="barh", color="blue")
plt.title("Frequency of Crimes by Crime Type")
plt.xlabel("ID")
plt.ylabel("Primary Type")
plt.show()
```
